```python
# get file to set up map
world = geopandas.read_file(
    geopandas.datasets.get_path('naturalearth_lowres')
)
world.head(10).sort_values(by='continent')
```

```python
pd.unique(world['name'])
```

```python
# combine the world and grouped_by_pos dataframes
combined_df_pos = pd.merge(grouped_by_pos, world, left_on="Code", right_on="iso_a3", how="inner")
combined_df_pos
```

```python
world_copy = world.copy(deep=True)
```

```python
# function for setting up the map on share of population with some formal education
# (countries with no examples are assigned zero)
def pos_func(world_cpy):
    if grouped_by_pos['Code'].isin([world_cpy]).any():
        row = grouped_by_pos[grouped_by_pos['Code'].isin([world_cpy])]
        pos = row["Share of population with some formal education, 1820-2020"].values[0]
        return pos
    else:
        return 0
```

```python
world_copy['pos'] = world_copy['iso_a3'].apply(pos_func)
```

```python
fig, ax = plt.subplots(figsize = (10,4), facecolor = plt.cm.Greys(.1))
fig.suptitle('Share of population with some formal\neducation for the years 2000-2020',
             fontsize = '15',  
             fontweight = 'bold')
ax.set_facecolor(plt.cm.Blues(.2))
world_copy.plot(column = 'pos',
           cmap = 'BuGn',
           ax = ax,
           legend = True)

plt.show()
```
